```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

torch.random.manual_seed(0)

model_name = "microsoft/Phi-3-mini-4k-instruct"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="cuda",     # всё на видеокарте
    torch_dtype="auto",    # сам выберет (обычно float16 на 30xx)
    # trust_remote_code=True,  # УБРАТЬ ЭТУ СТРОКУ
)

tokenizer = AutoTokenizer.from_pretrained(model_name)

messages = [
    {"role": "system", "content": "Ты ии ассистент - помощник"},
    {"role": "user", "content": "Можете ли вы предложить варианты сочетания бананов и драгонфрутов?"},
    {"role": "assistant", "content": "Конечно! Вот несколько способов есть бананы и драгонфруты вместе: 1. Смузи из бананов и драгонфрутов: Смешайте бананы и драгонфруты с небольшим количеством молока и меда. 2. Салат из бананов и драгонфрутов: Смешайте нарезанные бананы и драгонфруты с небольшим количеством лимонного сока и меда."},
    {"role": "user", "content": " "},
]

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.0,
    "do_sample": False,
}

output = pipe(messages, **generation_args)
print(output[0]["generated_text"])
```
